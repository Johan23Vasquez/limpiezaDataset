```python
import pandas as pd
import matplotlib.pyplot as plt
import pyarrow as pa
import pyarrow.parquet as pq
import dateparser
```

```python
archivo_mayo = "../data/001_raw/29-may-2026.csv"


may = pd.read_csv(archivo_mayo, encoding='ISO-8859-1',index_col=11)
```

```python
may.index = pd.to_datetime(may.index, format="mixed")
may.index
```

```python
columnas = may.columns
may = may.rename(columns={
    "Temperatura": "Temp",
    "Bulbo húmedo": "BH",
    "Punto de condensación": "PC"
    })
```

```python
may["Temp"] = may["Temp"].replace(" °C", "", regex=True)
may["PC"] = may["PC"].replace(" °C", "", regex=True)
may["BH"] = may["BH"].replace(" °C", "", regex=True)
may["% HR"] = may["% HR"].replace(" %", "", regex=True)
may["CO2"] = may["CO2"].replace(" ppm", "", regex=True)


may["Temp"] = pd.to_numeric(may["Temp"])
may["BH"] = pd.to_numeric(may["BH"])
may["PC"] = pd.to_numeric(may["PC"])
may["% HR"] = pd.to_numeric(may["% HR"])
may["CO2"] = pd.to_numeric(may["CO2"])
```

```python
may = may[["Temp", "BH", "PC", "% HR", "CO2"]]
```

```python
may_parquet = may.to_parquet("../data/002_intermediate/29_may_clean.parquet")
```
