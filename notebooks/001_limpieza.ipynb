```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pyarrow as pa
import pyarrow.parquet as pq
```

```python
archivo_mayo = "../data/raw/29-may-2026.csv"


may = pd.read_csv(archivo_mayo, encoding='ISO-8859-1',index_col=11, parse_dates=True)
```

```python
columnas = may.columns
may = may.rename(columns={
    "Temperatura": "Temp",
    "Bulbo húmedo": "BH",
    "Punto de condensación": "PC",
    "Sello de tiempo": "time"
    })
```

```python
may["Temp"] = may["Temp"].replace(" °C", "", regex=True)
may["PC"] = may["PC"].replace(" °C", "", regex=True)
may["BH"] = may["BH"].replace(" °C", "", regex=True)
may["% HR"] = may["% HR"].replace(" %", "", regex=True)
may["CO2"] = may["CO2"].replace(" ppm", "", regex=True)


may["Temp"] = pd.to_numeric(may["Temp"])
may["BH"] = pd.to_numeric(may["BH"])
may["PC"] = pd.to_numeric(may["PC"])
may["% HR"] = pd.to_numeric(may["% HR"])
may["CO2"] = pd.to_numeric(may["CO2"])
```

```python
may = may.loc[:, ["Lectura", "Temp", "BH", "PC", "% HR", "CO2"]]
may_index_nombre = may.rename_axis('Time')
```

```python
may_parquet = may_index_nombre.to_parquet("../data/processed/29_may_clean.parquet")
```
