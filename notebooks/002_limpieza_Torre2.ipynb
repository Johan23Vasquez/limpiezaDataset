```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
archivo_Torre2 = "../data/001_raw/002_Torre2.csv"
torre = pd.read_csv(archivo_Torre2, index_col=0, parse_dates=True)

archivo_mayo = "../data/002_intermediate/29_may_clean.parquet"
may = pd.read_parquet(archivo_mayo)
may.index
```

```python

torre.index = pd.to_datetime(torre.index, format="mixed").tz_localize(None)
torre
```

```python
torre_min = torre.copy()
may_min = may.copy()

torre_min.index = torre_min.index.floor("min")
may_min.index = may_min.index.floor("min")


torre_min = torre_min.groupby(level=0).mean()

concatenado = pd.concat(
    [torre_min, may_min],
    axis=1,
    join="inner"
)

concatenado
```

```python
may_torre = concatenado.to_parquet("../data/002_intermediate/may_torre.parquet")
```
